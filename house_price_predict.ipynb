```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('/Bengaluru_House_Data.csv')
```

```python
data2=data.drop(['area_type','society','balcony','availability'],axis='columns')
data2.head()
data3=data2.dropna(subset=['location','size'])#total value
data4=data3.loc[data3['bath'].isna(),['price','total_sqft']]
data4[['total_sqft']].head()
```

```python
for j,i in data4.iterrows():
    try:
      data4.at[j, 'total_sqft']=int((i['total_sqft'].split(' - '))[1])
    except ValueError:
      data4.at[j, 'total_sqft']=float((i['total_sqft'].split(' - '))[1])
    except:
      pass
data4.head()#bath null
```

```python
import re
x=data3[data3.total_sqft.apply(lambda X:True if re.search(r'[a-zA-Z]', X) else False)].total_sqft
ind=list(x.index)
#x=x.unique()
print(len(x),len(ind))
n=[]
s=[]
for i in x:
  match = re.match(r'(\d+(?:\.\d+)?)(.*)', i)
  n.append(float(match.group(1)))
  s.append((match.group(2)).lower())
a=list(zip(n,s))
print(a)
conversion = {
    'sq. meter': 10.7639,
    'perch': 272.25,
    'sq. yards': 9,
    'acres': 43560,
    'cents': 435.6,
    'guntha': 1089,
    'grounds': 2400
}
result = [round((i * conversion[j]),2) for i,j in a]
print(result)
#print(data3.at[648,'total_sqft'])
for i in range(len(ind)):
  data3.loc[ind[i],'total_sqft']=result[i]
data3.total_sqft.unique()
```

```python
import re
data5=data3[data3['bath'].notna()][['bath', 'price', 'total_sqft']]
for j,i in data5.iterrows():
    try:
      a=list(map(float,(i['total_sqft'].split(' - '))))
      data5.at[j, 'total_sqft']=(a[0]+a[1])/2
    except:
      data5.at[j, 'total_sqft']=float(i['total_sqft'])

data3.total_sqft.unique() #bath not null
```

```python
for j,i in data3.iterrows():
    try:
      a=list(map(float,(i['total_sqft'].split(' - '))))
      data3.at[j, 'total_sqft']=(a[0]+a[1])/2
    except:
      data3.at[j, 'total_sqft']=float(i['total_sqft'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(data5[['price','total_sqft']],data5['bath'])
pred = knn.predict(data4[['price','total_sqft']])
print(pred)
print(data4.index)
```

```python
j=0
for i in data4.index:
  data3.loc[i,'bath']=pred[j]
  j+=1
data3.head()
```

```python
data3['bed']=data3['size'].apply(lambda x: int(x.split(' ')[0]))
data3.head()
```

```python
data3['p_ft']=data3['price']*100000/data3['total_sqft']
data3.head()
```

```python
stat=data3.groupby('location')['location'].agg('count').sort_values(ascending=False)
stat
```

```python
stat2=stat[stat<=10]
stat2
```

```python
len(data3[(data3.total_sqft/data3.bed)<300])
```

```python
data3['location']=data3['location'].apply(lambda x: 'other' if x in stat2 else x)
len(data3.location.unique())
```

```python
data4=data3[(data3.total_sqft/data3.bed)>=300]
```

```python
data4.shape
#data4['p_ft'].describe()
```

```python
data5=pd.DataFrame()
for i,j in data4.groupby('location'):
  m=np.mean(j.p_ft)
  std=np.std(j.p_ft)
  ind=j[(j.p_ft>(m-std)) & (j.p_ft<=(m+std))]
  data5=pd.concat([data5,ind],ignore_index=True)
data5.shape
```

```python
ex=np.array([])
for i,j in data5.groupby('location'):
  bed_stats={}
  for k,l in j.groupby('bed'):
    bed_stats[k]={
        'mean':np.mean(l.p_ft)
        ,'std':np.std(l.p_ft)
        ,'count':len(l)
    }
    for k,l in j.groupby('bed'):
      stats=bed_stats.get(k-1)
      if stats and stats['count']>5:
        ex=np.append(ex,l[l.p_ft<stats['mean']].index.values)
data5.drop(ex,inplace=True)
data5.shape
```

```python
data5=data5[data5.bath<data5.bed+2]
```

```python
data6=data5.drop(['size','p_ft'],axis='columns')
data6.head()
```

```python
dummy=pd.get_dummies(data6.location)
```

```python
data7=pd.concat([data6.drop('location',axis='columns'),dummy.drop('other',axis='columns')],axis='columns')
data7.head()
```

```python
xtr,xts,ytr,yts=train_test_split(data7.drop('price',axis='columns'),data7.price,test_size=0.2,random_state=10)
```

```python
x=data7.drop('price',axis='columns')
y=data7.price
```

```python
from os import pipe
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('lasso', Lasso(alpha=0.1))
])
pipeline.fit(x, y)
pipeline.score(xts,yts)
```

```python
x.columns
```

```python
def predict_price(location,sqft,bath,bed):
    loc_index = np.where(x.columns==location)[0][0]
    x1 = np.zeros(len(x.columns))
    x1[0] = sqft
    x1[1] = bath
    x1[2] = bed
    if loc_index >= 0:
        x1[loc_index] = 1

    return pipeline.predict([x1])[0]
```
